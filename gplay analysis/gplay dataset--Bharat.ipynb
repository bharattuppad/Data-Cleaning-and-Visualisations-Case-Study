```python
def clean_size(x):
    if 'k' in x:
        x=x.replace('k','')
        return float(x)/1024
    elif 'M' in x:
        x=x.replace('M','')
        return float(x)
    elif 'Varies with device' in x:
        return np.nan
    else:
        return float(x)
    
```

```python
data['Size'] = data['Size'].map(clean_size)
```

```python
#Urban Property Area Loan Status
plt.style.use('ggplot')

parameter = [t_f, t_p]
labels = ['Free App', 'Paid App']
explode = [0, 0.35]

plt.pie(parameter, labels=labels, explode=explode, colors=['green','red'], shadow=True,
        startangle=0, autopct='%1.1f%%', wedgeprops={'edgecolor':'black'})

plt.title('Type of Apps with Rating > 4.0')
plt.show()
```
